```python
#outlier removal
def outlier_removal(data,feature):
    Q1=data[feature].quantile(.25)
    Q3=data[feature].quantile(.75)
    IQR=Q3-Q1
    LL=Q1-(1.5*IQR)
    UL=Q3+(1.5*IQR)
    data=data[(data[feature]>LL) & (data[feature]<UL)]
    return data
Ecom_Customer_data=outlier_removal(Ecom_Customer_data,"Length of Membership")
Ecom_Customer_data=outlier_removal(Ecom_Customer_data,"Avg. Session Length")
Ecom_Customer_data=outlier_removal(Ecom_Customer_data,"Time on App")
Ecom_Customer_data=outlier_removal(Ecom_Customer_data,"Time on Website")
Ecom_Customer_data=outlier_removal(Ecom_Customer_data,"Length of Membership")
```

```python
# Outlier Analysis
fig, axs = plt.subplots(5, figsize = (5,10))
plt1 = sns.boxplot(Ecom_Customer_data['Avg. Session Length'], ax = axs[0])
plt2 = sns.boxplot(Ecom_Customer_data['Time on App'], ax = axs[1])
plt3 = sns.boxplot(Ecom_Customer_data['Time on Website'], ax = axs[2])
plt4 = sns.boxplot(Ecom_Customer_data['Length of Membership'], ax = axs[3])
plt5 = sns.boxplot(Ecom_Customer_data['Yearly Amount Spent'], ax = axs[4])
plt.tight_layout()
```

# Model Selection and Evaluation

```python
#Importing the train_test_split to separate the training and testing data
from sklearn.model_selection import train_test_split 
```

```python
#Here as "email","avatar" and "address" is not effecting the "Yearly amount spent" so we are droping those columns
#We are assigning 'Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership' to "X" value
#We are assigning "Yearly Amount Spent" to "Y" value
X=Ecom_Customer_data.drop(["Yearly Amount Spent","Email","Address","Avatar"],axis=1)
Y=Ecom_Customer_data["Yearly Amount Spent"]
```

```python
#Printing the first five rows of X value
X.head()
```

```python
#Printing the first five rows of Y value
Y.head()
```

```python
#Checking the shape of total dataset
Ecom_Customer_data.shape
```

```python
#Checking the shape of X value
X.shape
```

```python
#Checking the shape of Y value
Y.shape
```

```python
#Spliting the data and assigning to X_train,X_test,Y_train,Y_test variables
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#Checking the shape of training dataset
print(X_train.shape,Y_train.shape)
```

```python
#Checking the shape of test dataset
print(X_test.shape,Y_test.shape)
```

```python
#Importing the models and necessary modules
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Storing the required models into a list
models=[LinearRegression(),Lasso(),Ridge()]
```

```python
for model in models:
    #Traing the model
    m=model.fit(X_train,Y_train)
    
    c=m.coef_
    #Testing the model
    Y_pred=m.predict(X_test)
    
    #Training Accuracy
    train_accuracy=model.score(X_train,Y_train)
    
    #Testing Accuracy
    testing_accuracy=model.score(X_test,Y_test)
    
    #Calculating the mean_square_error
    mean_square_error=mean_squared_error(Y_pred,Y_test)
    
    #Caculating the r2_score
    r2=r2_score(Y_pred,Y_test)
    
    x=str(model)
    print("Model : ",x[:-2].upper())
    print(" Mean_square_error : ",mean_square_error)
    print(" R2_score : ",r2)
    print(" Coefficients of the model : ",X.columns)
    print(" Coefficients of the model : ",c)
    print("ACCURACY OF THE MODEL WITH TRAINING DATA SET IS ",train_accuracy*100)
    print("ACCURACY OF THE MODEL WITH TESTING DATA SET IS ",testing_accuracy*100)
    print("*************************************************************************************")
   
```
